### 1) Import packages

```python
import numpy as np
import matplotlib.pyplot as plt
```

### 2) Plotting

We plot below (@fig-histogram) the histogram of 1000 samples from a normal distribution.

```python
#| label: fig-histogram
#| fig-cap: Histogram of X
x = np.random.normal(size=1000)
plt.figure(figsize=(10, 6))
plt.hist(x, density=True, bins=30)
plt.ylabel('Probability')
plt.xlabel('X')
plt.show()
```
